```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
zp_data = pd.read_excel('/content/sample_data/zp_srmes_rab.xlsx', index_col=0)
zp_data.head()
```

```python
infl_data = pd.read_excel('/content/sample_data/infl_rab.xlsx', index_col=0)
infl_data.head(13)
```

```python
mask1 = zp_data.index == 'рыболовство и рыбоводство'
mask2 = infl_data.index == 'Всего'
mask3 = zp_data.index == 'образование'
mask4 = zp_data.index == 'добыча металлических руд'

plt.figure(figsize=(5, 5))
plt.plot(zp_data.columns, zp_data[mask1].values[0])
plt.plot(zp_data.columns, zp_data[mask3].values[0])
plt.plot(zp_data.columns, zp_data[mask4].values[0])
plt.title("ЗП без учета инфляции")
plt.legend(['Рыболовство и рыбоводство','Образование','Добыча металлических руд'])
plt.show()

plt.figure(figsize=(5, 5))
plt.bar(infl_data.columns, infl_data[mask2].values[0],label='Величина прибыли')
plt.title("Инфляция")
plt.legend(['Инфляция'])
plt.show()

plt.figure(figsize=(5, 5))
plt.plot(zp_data.columns, zp_data[mask1].values[0])
plt.plot(infl_data.columns, (1-infl_data[mask2].values[0]/100)*zp_data[mask1].values[0])
plt.bar(zp_data.columns,zp_data[mask1].values[0]-(1-infl_data[mask2].values[0]/100)*zp_data[mask1].values[0])
plt.title("Рыболовство и рыбоводство")
plt.legend(['ЗП без учета инфляции','ЗП с учетом инфляции','delta'])
plt.show()

plt.figure(figsize=(5, 5))
plt.plot(zp_data.columns, zp_data[mask3].values[0])
plt.plot(infl_data.columns, (1-infl_data[mask2].values[0]/100)*zp_data[mask3].values[0])
plt.bar(zp_data.columns,zp_data[mask3].values[0]-(1-infl_data[mask2].values[0]/100)*zp_data[mask3].values[0])
plt.title("Образование")
plt.legend(['ЗП без учета инфляции','ЗП с учетом инфляции','delta'])
plt.show()

plt.figure(figsize=(5, 5))
plt.plot(zp_data.columns, zp_data[mask4].values[0])
plt.plot(infl_data.columns, (1-infl_data[mask2].values[0]/100)*zp_data[mask4].values[0])
plt.bar(zp_data.columns,zp_data[mask4].values[0]-(1-infl_data[mask2].values[0]/100)*zp_data[mask4].values[0])
plt.title("Добыча металлических руд")
plt.legend(['ЗП без учета инфляции','ЗП с учетом инфляции','delta'])
plt.show()
```

Вывод:
При анализе представленных данных было установлено снижение инфляции и рост среднемесячных зарплат.  Повышение зарплат обуславливает увеличение "разрыва" их номинальных и реальных значений.
